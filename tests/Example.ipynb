```python
from fluxKit import bap 
from bap import Image
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np
```

# Basic blacklight and AthenaMC Spectra Imaging

```python
im = Image('../example_data/exampleFile.npz')
```

```python
mc = bap.MCSpec('/home/tegan/Research/blacklightAthenaProcessing/example_data/',32,50,1,1000,outPath='/home/tegan/Research/blacklightAthenaProcessing/example_output/')
```

```python
fig,ax = plt.subplots()
image = bap.plot_spectra(im,ax,MC_spec=mc,labels=['free-free 1e7K blacklight'],MC_labels=['1e6K(?) Compton MC'])
plt.yscale('log')
plt.xscale('log')
plt.show()
```

# blacklight Image Production

```python
fig,ax = plt.subplots()
divider = make_axes_locatable(ax)
#cax = divider.append_axes('right', size='1%', pad=0.05)
image = bap.plot_image(im,'I',axes='rg',logc=True,cmin=1e-6,cmax=1e-1)
#fig.colorbar(image, cax=cax, orientation='vertical')
color = bap.colorbar(ax,image)
ax.set_title('Intensity')
plt.show()
```

```python
fig,ax = plt.subplots()
divider = make_axes_locatable(ax)
#cax = divider.append_axes('right', size='1%', pad=0.05)
image = bap.plot_image(im,'tau',axes='cm',logc=True,cmin=1e-6,cmax=1e-1)
#fig.colorbar(image, cax=cax, orientation='vertical')
color = bap.colorbar(ax,image)
ax.set_title('$\\tau$')
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,6))
image = bap.plot_spectra(im,ax[0],MC_spec=mc,labels=['free-free 1e7K blacklight'],MC_labels=['1e6K(?) Compton MC'])
ax[0].set_yscale('log')
ax[0].set_xscale('log')
divider = make_axes_locatable(ax[1])
#cax = divider.append_axes('right', size='1%', pad=0.05)
image = bap.plot_image(im,'I',axes='cm',logc=True,cmin=1e-6,cmax=1e-1)
#fig.colorbar(image, cax=cax, orientation='vertical')
color = bap.colorbar(ax[1],image)
ax[1].set_title('$\\tau$')
plt.show()
```
